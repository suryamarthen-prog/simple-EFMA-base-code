## Objectives

### Step 0: Importing Libraries

```python
# Importing Necessary Libraries
from ema_workbench import (
    ema_logging,
    perform_experiments,
    RealParameter,
    ScalarOutcome
)

from ema_workbench.connectors.excel import ExcelModel

import matplotlib.pyplot as plt
import seaborn as sns

ema_logging.log_to_stderr(ema_logging.INFO)
```

### Step 1: Performing The Experiments

```python
# Connect with your Excel and select which sheet contain your parameters to track and output
# Model File (R)
model = ExcelModel("ModelName", wd="models", model_file ="Excel filename.xlsx")   # have your Excel inside a folder called "models"
model.default_sheet = "SheetName"

# Define Uncertainties (X)
model.uncertainties = [RealParameter("Parameter 1", 0, 100),   # ("refer to a cell", min value, max value)
                       RealParameter("Parameter 2", -1, 1),
                       RealParameter("Parameter 3", -100, 20)
                       ] 

# Define to Outcomes (M)
model.outcomes = [ScalarOutcome("npv"),   # refer to a cell
                  ScalarOutcome("irr")
                  ] 
```

#### Step 1.5: Run Experiments

```python
# Run Experiments
experiments, outcomes = perform_experiments(model, scenarios = 50000)   # change 50000 to as many iteration you need
```

### Step 2: Analyze The Output

#### a. Feature Scoring
To interpret the output, the higher the number means it has higher influence

```python
from ema_workbench.analysis import feature_scoring
x = experiments
y = outcomes

fs = feature_scoring.get_feature_scores_all(x, y)
sns.heatmap(fs, cmap="viridis", annot=True)
plt.show()
```

#### Dimensional Stacking
To interpret the output, the lighter to yellow color means that combination of uncertainties produce more robust outcome

```python
# this cell only produce outcome for NPV indicator, copy as many as needed
from ema_workbench.analysis import dimensional_stacking

x = experiments
y = outcomes["npv"] > 0   # copy this cell to have another outcome to be analyzed; change "> 0 " to be your desired parameter
dimensional_stacking.create_pivot_plot(x, y, 2, nbins=3, bin_labels=True)   # experiment with nbins number to match your needs, higher means more bins with more detailed analysis
plt.show()
```
